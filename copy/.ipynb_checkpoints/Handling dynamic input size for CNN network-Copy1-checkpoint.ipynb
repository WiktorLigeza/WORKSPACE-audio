```python
from pydub import AudioSegment
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
import pydub
```

## Raw mfcc

```python
def raw_mfcc(path):
    X, sample_rate = librosa.load(path,res_type='kaiser_fast')

    # mfcc 
    mfccs = librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40)
    plt.figure(figsize=(17.5, 5))
    librosa.display.specshow(mfccs, 
                             x_axis="time", 
                             sr=sample_rate)
    plt.colorbar(format="%+2.f")
    plt.show()
    
    return mfccs.T
```

```python
audio_1 = "all_actors/03-01-01-01-01-01-01.wav"
audio_2 = "all_actors/03-01-01-01-01-01-02.wav"
raw_1 = raw_mfcc(audio_1)
raw_2 = raw_mfcc(audio_2)
```

## Padding method

```python
def get_mfcc_2D_with_audio_padding(path, duration_s = 7, sample_rate = 22050, show=False):
    # pad
    audio = pydub.AudioSegment.silent(duration=(duration_s*1000))    # The length you want in ms 
    audio = audio.overlay(pydub.AudioSegment.from_wav(path).set_frame_rate(22050))
    raw = audio.split_to_mono()[0].get_array_of_samples() 
    raw_np = np.array(raw).astype('float64')
    
    # get mfcc
    mfccs = librosa.feature.mfcc(y=raw_np, sr=sample_rate, n_mfcc=40)
    
    if show:
        plt.figure(figsize=(17.5, 5))
        librosa.display.specshow(mfccs, 
                                 x_axis="time", 
                                 sr=sample_rate)
        plt.colorbar(format="%+2.f")
        plt.show()
    return mfccs.T

    
```

```python
pad_1 = get_mfcc_2D_with_audio_padding(path = audio_1, show= True)
pad_2 = get_mfcc_2D_with_audio_padding(path = audio_2, show= True)
```

# CONCLUSION

the size of padded mfcc spectogram is const. -> can be used in CNN network

```python
print("raw mfccs shapes: 1-{}, 2-{}".format(raw_1.shape, raw_2.shape))
print("padded mfccs shapes: 1-{}, 2-{}".format(pad_1.shape, pad_2.shape))
```

```python
max_len_audio = "toronto_ds/conc_data/03-02-02-02-02-01-22.wav"
min_len_audio = "toronto_ds/conc_data/95-41-06-43-13-16-83.wav"

pad_1 = get_mfcc_2D_with_audio_padding(path=max_len_audio, show= True)
pad_2 = get_mfcc_2D_with_audio_padding(path=min_len_audio, show= True)

print("padded mfccs shapes: max-{}, min-{}".format(pad_1.shape, pad_2.shape))
```

```python
X1, sr1 = librosa.load(max_len_audio,
                        res_type='kaiser_fast')

X2, sr2 = librosa.load(min_len_audio,
                        res_type='kaiser_fast')
    
stft1 = librosa.stft(y=X1)
mfccs1 = librosa.feature.mfcc(y=X1, sr=sr1, n_mfcc=40)

stft2 = librosa.stft(y=X2)
mfccs2 = librosa.feature.mfcc(y=X2, sr=sr2, n_mfcc=40)

Y_log_scale1 = librosa.power_to_db(stft1)
Y_log_scale2 = librosa.power_to_db(stft2)

######################### plots 1 #########################

plt.figure(figsize=(17.5, 5))
librosa.display.specshow(Y_log_scale1, 
                         x_axis="time", 
                         sr=sr1)
plt.colorbar(format="%+2.f")
plt.show()

plt.figure(figsize=(17.5, 5))
librosa.display.specshow(mfccs1, 
                         x_axis="time", 
                         sr=sr1)
plt.colorbar(format="%+2.f")
plt.show()

###########################################################
```
