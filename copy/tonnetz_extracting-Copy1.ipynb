```python
import scipy
from spafe.features.gfcc import gfcc
import os
import glob
import joblib
import numpy as np
from tqdm import tqdm
import librosa
from librosa import display
import wavio

import scipy.io.wavfile
import spafe.utils.vis as vis
from spafe.features.gfcc import gfcc
```

```python
def extract(audio_file):
    name=os.path.basename(audio_file)
    name=os.path.splitext(name)[0]
    label = int(name[7:8]) - 1
    
    return label, audio_file
    
    
def return_feats(dir_name):
    X_ = []
    y_ = []
    err_cnt = 0
    err_list = []
    exeption_list = []
    
    #progression bar 
    file_count = len(dir_name)
    bar = tqdm(desc="proccessing files", total=file_count)
    tot = 0
    
    for file in dir_name:
        bar.update(1)
        tot += 1
        label, audio_file = extract(file)
        
        try:
            signal, sample_rate = librosa.load(audio_file, res_type='kaiser_fast')
            tonnetz = np.mean(librosa.feature.tonnetz(y=signal, sr=sample_rate).T,axis=0) 
            
            X_.append(tonnetz)
            y_.append(label)
            
        except Exception as e:
            err_cnt += 1
            err_list.append(audio_file)
            exeption_list.append(str(e))
            print("{}.error at:".format(err_cnt), audio_file)
            
        
    return X_, y_, err_cnt, err_list, exeption_list
```

```python
dir_name = glob.glob("toronto_ds/conc_data/*.wav")

X, y, _, _, _ = return_feats(dir_name)
X_name, y_name = 'tonnetz_feats/X.joblib', 'tonnetz_feats/y.joblib'

joblib.dump(np.float64(X), X_name)
joblib.dump(y, y_name)
```
